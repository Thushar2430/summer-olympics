```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('summer.csv')
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
print(f"Olympics is held so far in {len(df['City'].unique())} cities")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
    df1=df[df['Sport']==sport]
    data.append([sport, len(df1[df1['Medal']=='Gold'])])
ax=pd.DataFrame(data, columns=['Sport','medals']).sort_values(by='medals', ascending=False).head().plot(x='Sport',y='medals', kind='bar', figsize=(5,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
ax=pd.DataFrame(data, columns=(['Sport','Medals'])).sort_values(by='Medals', ascending=False).head().plot(x='Sport', y='Medals', kind='bar', figsize=(5,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for athlete in df['Athlete'].unique():
    data.append([athlete,len(df[df['Athlete']==athlete])])
ax=pd.DataFrame(data, columns=(['Athlete','Medals'])).sort_values(by='Medals', ascending=False).head().plot(x='Athlete', y='Medals', kind='bar', figsize=(5,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data=[]
for athlete in df['Athlete'].unique():
    df1=df[df['Athlete']==athlete]
    data.append([athlete,len(df1[df1['Medal']=='Gold'])])
ax=pd.DataFrame(data, columns=(['Athlete','Medals'])).sort_values(by='Medals', ascending=False).head().plot(x='Athlete', y='Medals', kind='bar', figsize=(6,6))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df1=df[df['Country']=='IND']
df2=df1[df1['Medal']=='Gold']
yr=[]
for i in df2['Year']:
    yr.append(i)
print(f'India won first Gold medal in {min(yr)}.')
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for event in df['Event'].unique():
    df1=df[df['Event']==event]
    data.append([event,len(df1)])
ax=pd.DataFrame(data, columns=['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x='Event', y='Players', kind='bar', figsize=(5,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
    df1=df[df['Sport']==sport]
    df2=df1[df1['Gender']=='Women']
    data.append([sport, len(df2[df2['Medal']=='Gold'])])
ax=pd.DataFrame(data, columns=['Sport','medals']).sort_values(by='medals', ascending=False).head().plot(x='Sport',y='medals', kind='bar', figsize=(5,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
